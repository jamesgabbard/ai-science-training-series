```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Timings
print("Mean time per epoch for  1 GPUs: %1.3f s" % (np.mean([1.535, 1.528, 1.529, 1.532, 1.542])))
print("Mean time per epoch for  2 GPUs: %1.3f s" % (np.mean([1.609, 1.593, 1.602, 1.604, 1.660])))
print("Mean time per epoch for  4 GPUs: %1.3f s" % (np.mean([0.660, 0.708, 0.719, 0.771, 0.661])))
print("Mean time per epoch for  8 GPUs: %1.3f s" % (np.mean([0.557, 0.564, 0.558, 0.560, 0.544])))
print("Mean time per epoch for 16 GPUs: %1.3f s" % (np.mean([0.507, 0.507, 0.510, 0.504, 0.503])))
```

```python
# Plot
for ranks in [1, 2, 4, 8, 16]:
    data = np.loadtxt("plot_data/hw_loss_accuracy_%d.txt" % ranks, delimiter=' ')

    fig, (ax1, ax2) = plt.subplots(ncols = 2)
    
    ax1.plot(data[1:, 2], "-")
    ax1.set_xlabel("Steps")
    ax1.set_ylabel("Loss")
    ax1.set_title("Loss History, %d GPUs" % ranks)

    ax2.plot(data[1:, 3], "-")
    ax2.set_xlabel("Steps")
    ax2.set_ylabel("Accuracy")
    ax2.set_title("Accuracy History, %d GPUs" % ranks)

    fig.set_tight_layout(True)
    fig.savefig("accuracy_%d.pdf" % ranks)
```
